```python
import pandas as pd # Biblioteca para processamento de dados
import numpy as np # Biblioteca para Álgebra Linear
import pickle # Biblioteca para salvar/carregar modelos de machine learning

# Bibliotecas para pré-processamento e transformações dos dados
from sklearn import preprocessing
from sklearn import decomposition

# separar dados em teste e treino
from sklearn.model_selection import train_test_split

# Biblioteca para machine learning usando multi layers Perceptron
from sklearn.neural_network import MLPClassifier

# Bibliotecas para avaliar o modelo
from sklearn import metrics
from sklearn.metrics import classification_report

# tratar warnings
import warnings

warnings.filterwarnings('ignore')
```

```python
# carregando os dados
file_path = "../Dados/train.csv"
file_path_teste = "../Dados/test.csv"

dataSet = pd.read_csv(file_path)
dataSet_teste = pd.read_csv(file_path_teste)

dataSet.head()
```

```python
# visualizando o 'formato' dos dados
dataSet.describe()
```

```python
# verificando os tipos das colunas
dataSet.info()
```

```python
# verificando se há valores NA
dataSet.isna().sum().sum()
```

```python
# normlizando os dados tanto para treino quanto teste
min_max_scaler = preprocessing.MinMaxScaler()
dataSet2 = pd.DataFrame(min_max_scaler.fit_transform(dataSet.drop(['label'], axis = 1)))

dataSet_teste2 = pd.DataFrame(min_max_scaler.fit_transform(dataSet_teste))
```

```python
dataSet2['label'] = dataSet['label']
```

```python
# visualizando o novo 'formato' dos dados
dataSet2.describe()
```

```python
# separando o dataset em treino e teste
train, test = train_test_split(dataSet2, test_size = 0.3)
```

```python
# https://scikit-learn.org/stable/modules/neural_networks_supervised.html

# modelo - MLP(redes neurais)
modelo_MLP = MLPClassifier(hidden_layer_sizes = (800, 100),
                           alpha = 1e-4,
                           random_state = 101,
                           solver='adam')


modelo_MLP.fit(train.drop(['label'], axis = 1), train['label'])
```

```python
previsoes_MLP = modelo_MLP.predict(test.drop(['label'], axis = 1))

metrics.accuracy_score(test['label'], previsoes_MLP)*100
```

```python
# Avaliando o modelo - usando métricas de classificação
print(classification_report(test['label'], previsoes_MLP))
```
